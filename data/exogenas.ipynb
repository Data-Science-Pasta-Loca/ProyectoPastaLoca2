```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define el símbolo del ticker para el par de divisas (GBP/EUR)
ticker = 'GBPEUR=X'

# Obtener datos históricos
data = yf.download(ticker, start='2020-10-20', end='2020-11-07')

# Mantener solo la columna 'Close' (Cierre)
exchange_rate = data[['Close']]

# Restablecer el índice para que 'Date' (Fecha) sea una columna
exchange_rate = exchange_rate.reset_index()

# Renombrar las columnas a 'Date' (Fecha) y 'GBP to EUR' (GBP a EUR)
exchange_rate.columns = ['Date', 'GBP to EUR']

# Mostrar el DataFrame
display(exchange_rate.tail(10))

# Define el símbolo del ticker para BTC/GBP
ticker = 'BTC-GBP'

# Obtener datos históricos
data4 = yf.download(ticker, start='2020-10-20', end='2020-11-07')

# Mantener solo la columna 'Close' (Cierre)
btc_gbp_data = data4[['Close']]

# Restablecer el índice para que 'Date' (Fecha) sea una columna
btc_gbp_data = btc_gbp_data.reset_index()

# Renombrar las columnas a 'Date' (Fecha) y 'BTC to GBP' (BTC a GBP)
btc_gbp_data.columns = ['Date', 'BTC to GBP']

# Mostrar el DataFrame
display(btc_gbp_data.tail(10))

#df_jo = pm.df('df_jo')
#df_jo['created_at_d'] = pd.to_datetime(df_jo['created_at'])
#df_jo['created_at_d'] = df_jo['created_at_d'].dt.date

daily = pd.DataFrame({'Date': pd.date_range(start='2019-12-01', end='2020-11-04', freq='D')})
daily['Date'] = pd.to_datetime(daily['Date'])
daily['Date'] = daily['Date'].dt.date


exchange_rate['Date'] = pd.to_datetime(exchange_rate['Date'],format='ISO8601')
exchange_rate['Date'] = exchange_rate['Date'].dt.date
daily = pd.merge(daily, exchange_rate, left_on='Date', right_on='Date', how ="left") #inner
daily['GBP to EUR'] = daily['GBP to EUR'].fillna(0)



btc_gbp_data['Date'] = pd.to_datetime(btc_gbp_data['Date'],format='ISO8601')
btc_gbp_data['Date'] = btc_gbp_data['Date'].dt.date
daily = pd.merge(daily, btc_gbp_data, left_on='Date', right_on='Date', how ="left") #inner
daily['BTC to GBP'] = daily['BTC to GBP'].fillna(0)


#daily = daily.drop(columns=['Date_x','Date_y'])


# Combine the two DataFrames by the 'Date' column
#divisa_exogenas = pd.merge(btc_gbp_data, exchange_rate, on='Date', how='left')

display(daily.tail(15))
```
