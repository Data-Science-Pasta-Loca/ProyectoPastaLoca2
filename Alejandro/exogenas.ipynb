```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the ticker symbol for the currency pair (GBP/EUR)
ticker = 'GBPEUR=X'

# Fetch historical data
data = yf.download(ticker, start='2019-11-01', end='2020-10-31')

# Keep only the 'Close' column
df = data[['Close']]

# Make 'Date' the index (this should already be done automatically)
df = df.reset_index()  # Convert 'Date' from index to column first
df.set_index('Date', inplace=True)  # Set 'Date' as the new index

# Display the resulting DataFrame
df
```

```python

# Plot the 'Close' column
plt.figure(figsize=(10, 6))  # Set figure size
plt.plot(df.index, df['Close'], label='GBP/EUR Close', color='blue')  # Plot the data

# Add labels and title
plt.title('GBP/EUR Exchange Rate', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Close Price', fontsize=12)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)
# Display grid
plt.grid(True)

# Show the plot
plt.legend()
plt.tight_layout()  # Adjust layout to fit everything
plt.show()
```

```python
import pandas as pd

# Define the data
data = {
    'Date': ['Dec 19', 'Jan 20', 'Feb 20', 'Mar 20', 'Apr 20', 'May 20', 'Jun 20', 'Jul 20', 'Aug 20', 'Sep 20', 
             'Oct 20'],
    'Inflation (%)': [1.3, 1.8, 1.7, 1.5, 0.8, 0.5, 0.6, 1.0, 0.2, 0.5, 0.7]
}

#Fuente: https://www.statista.com/statistics/306648/inflation-rate-consumer-price-index-cpi-united-kingdom-uk/

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Inflation (%)'], marker='o', color='b', linestyle='-', linewidth=2, markersize=6)

# Adding titles and labels
plt.title('Inflation', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Value (%)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the grid
plt.grid(True)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
